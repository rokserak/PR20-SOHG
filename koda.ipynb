```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')

inflacija_df = pd.read_csv('inflacija.csv', sep=',', header=1, encoding='windows-1250')
gospodinstva_df=pd.read_csv('gospodinstva.csv', sep=',',header=1)

inflacija_skupaj = inflacija_df.loc[0][13:] - 100

inflacija_jan = [inflacija_df.loc[0][i] / 100 for i in range(13, inflacija_df.loc[0][13:].size, 12)]

def calc_inflation(i, j):
    inflation = 1
    for k in range(i, j + 1):
        inflation *= inflacija_jan[k]
    return inflation * 100

inflacija_jan = [calc_inflation(0, i) - 100 for i in range(len(inflacija_jan))]
cena_z_inflacijo = [
    float(cena) - float(cena) * inflacija / 100 for cena, inflacija
                                                in zip(gospodinstva_df.loc[1][7:-1], inflacija_jan)
]
# add for year 1999 with no inflation included
cena_z_inflacijo.insert(0, gospodinstva_df.loc[1][6])
#cena_z_inflacijo

```

```python
inflacija_df = pd.read_csv('inflacija.csv', sep=',', header=1, encoding='windows-1250')
inflacija_2008_naprej = inflacija_df.loc[0][97:]
```

```python
letna_inflacija_2008_naprej = [inflacija_df.loc[0][i] / 100 for i in range(13, inflacija_df.loc[0][97:].size, 12)]

def calc_inflation(i, j):
    inflation = 1
    for k in range(i, j + 1):
        inflation *= letna_inflacija_2008_naprej[k]
    return inflation * 100

letna_inflacija_2008_naprej = [calc_inflation(0, i) - 100 for i in range(len(letna_inflacija_2008_naprej) - 1)]

leta = [str(leto) for leto in range(2008, 2019)]
```

```python
plt.figure(figsize=(15, 8))
plt.plot(leta, letna_inflacija_2008_naprej)
plt.title('Inflacija od leta 2008 do 2019')
plt.xlabel('Leto')
plt.ylabel('Inflacija (%)')
plt.show()
```
