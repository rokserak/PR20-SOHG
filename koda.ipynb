```python
import scipy.cluster.hierarchy as sch
L = sch.linkage(podatkiBrezInflacije, method="average", metric="cosine")
plt.figure(figsize=(15, 7))
#labels = [row["name"].value for row in data]
D      = sch.dendrogram(L, labels=imena, orientation="left", leaf_font_size=15)
plt.xlabel("Razdalja")
plt.title('Gručenje dobrin glede na porabo skozi leta')
plt.tight_layout()
plt.savefig('5.png')
plt.show()

#Kosinusna razdalja (metric="cosine")
#Predstavlja kosinus kota med vektorjema  𝑥⃗   in  𝑦⃗   - manjši kot pomeni večjo podobnost. Uporabna za primerjavo podobnosti med vektorji, neupoštevajoč absolutnih #velikosti.
```

```python
inflacija_df = pd.read_csv('inflacija.csv', sep=',', header=1, encoding='windows-1250')
inflacija_2008_naprej = inflacija_df.loc[0][97:]
```

```python
letna_inflacija_2008_naprej = [inflacija_df.loc[0][i] / 100 for i in range(13, inflacija_df.loc[0][97:].size, 12)]

def calc_inflation(i, j):
    inflation = 1
    for k in range(i, j + 1):
        inflation *= letna_inflacija_2008_naprej[k]
    return inflation * 100

letna_inflacija_2008_naprej = [calc_inflation(0, i) - 100 for i in range(len(letna_inflacija_2008_naprej) - 1)]

leta = [str(leto) for leto in range(2008, 2019)]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(leta, letna_inflacija_2008_naprej)
plt.title('Inflacija od leta 2008 do 2019')
plt.xlabel('Leto')
plt.ylabel('Inflacija (%)')
plt.show()
```
